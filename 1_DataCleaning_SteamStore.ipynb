```python
def find_uniques(df,feature='feature',sep=';'):
    df[feature] = df[feature].apply(lambda x: x.split(sep))

    genre_list = []

    df[feature].apply(lambda x: [genre_list.append(g) for g in x if g not in genre_list])

    return genre_list
```

```python
genres_list = find_uniques(df1, 'genres', ';')
genres_list, len(genres_list)
```

```python
categories_list = find_uniques(df1,'categories',';')
categories_list, len(categories_list)
```

```python
# create genres and categories one hot encode
df1 = df1.drop('genres', 1).join(df1.genres.str.join('|').str.get_dummies())
```

```python
df1 = df1.drop('categories', 1).join(df1.categories.str.join('|').str.get_dummies())
```

```python
# drop some unnecessary columns
df1.drop(columns=['english','steamspy_tags','owners'], inplace=True)
```
